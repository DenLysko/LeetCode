```python
class Solution:
    def mergeAlternately(self, word1: str, word2: str) -> str:
        chars1 = [char for char in word1]
        chars2 = [char for char in word2]
        result = ""
        min_length = min(len(chars1), len(chars2))
        for i in range(min_length):
            result += chars1[i]
            result += chars2[i]
        
        if len(chars1) > len(chars2):
            result += ''.join(chars1[min_length:])
        else:
            result += ''.join(chars2[min_length:])

        return result
```

```python
class Solution:
    def gcdOfStrings(self, str1: str, str2: str) -> str:
        min_str = ""
        max_str = ""
        if len(str1) > len(str2):
            min_str = str2
            max_str = str1
        elif len(str1) < len(str2):
            min_str = str1
            max_str = str2
        elif str1 == str2:
            return str1
        else:
            return ""
        
        min_length = min(len(str1), len(str2))
        max_length = max(len(str1), len(str2))

        # if max_length % min_length != 0:
            # return ""
        
        counter = 0
        result = ""    
        common_divider = ""
        
        while (counter < min_length):
            common_divider += min_str[counter]
            counter += 1
            if (min_length % len(common_divider) == 0 and max_length % len(common_divider) == 0):
                ratio1 = int(min_length / len(common_divider))
                ratio2 = int(max_length / len(common_divider))
                if (max_str == common_divider * ratio2 and min_str == common_divider * ratio1):
                    result = common_divider

        # print(result)
        return result
```

```python
class Solution:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
        maximum_value_in_initial_list = 0
        for i in range(len(candies)):
            if candies[i] > maximum_value_in_initial_list:
                maximum_value_in_initial_list = candies[i]
        
        result = []
        for i in range(len(candies)):
            if candies[i] + extraCandies >= maximum_value_in_initial_list:
                result.append(True)
            else:
                result.append(False)
        
        return result
```

```python
class Solution:
    def reverseVowels(self, s: str) -> str:
        vowels_in_string = []
        indexes = []
        chars = [char for char in s]
        all_vowels = ['a', 'e', 'i', 'o', 'u']
        for i in range(len(chars)):
            if chars[i].lower() in all_vowels:
                vowels_in_string.append(chars[i])
                indexes.append(i)
        
        for i in range(len(vowels_in_string)):
            chars[indexes[len(vowels_in_string) - 1 - i]] = vowels_in_string[i]
        
        return ''.join(chars)
```

```python
class Solution:
    def largestAltitude(self, gain: List[int]) -> int:
        prefix_sum = 0
        max_height = 0
        for i in range(len(gain)):
            prefix_sum += gain[i]
            if prefix_sum > max_height:
                max_height = prefix_sum
        return max_height
```

```python
class Solution:
    def pivotIndex(self, nums: List[int]) -> int:
        summ = sum(nums)
        pre_sum = 0
        for i in range(len(nums)):
            if pre_sum == (summ - nums[i] - pre_sum):
                return i
            pre_sum += nums[i]
        return -1
```

```python
class Solution:
    def findDifference(self, nums1: List[int], nums2: List[int]) -> List[List[int]]:
        set1 = set(nums1)
        set2 = set(nums2)
        arr1 = []
        arr2 = []

        for item in set1:
            if not set2.__contains__(item):
                arr1.append(item)
        
        for item in set2:
            if not set1.__contains__(item):
                arr2.append(item)
        
        return [arr1, arr2]
```

```python
class Solution:
    def uniqueOccurrences(self, arr: List[int]) -> bool:
        dict = {}
        for i in arr:
            if i in dict:
                dict[i] += 1
            else:
                dict[i] = 1
        
        return len(set(dict.values())) == len(dict.values())
```

```python
class Solution:
    def findMaxAverage(self, nums: List[int], k: int) -> float:
        left_index = 0
        right_index = k - 1
        tmp_sum = sum(nums[:k])
        max_tmp_sum = tmp_sum
        while (right_index < len(nums) - 1):
            right_index += 1
            tmp_sum = tmp_sum - nums[left_index] + nums[right_index]
            if tmp_sum > max_tmp_sum:
                max_tmp_sum = tmp_sum
            left_index += 1
        return max_tmp_sum / k
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        stack = []
        current_node = head
        if head is None:
            return None
        stack.append(current_node)
        tmp_node = current_node
        while (not current_node.next == None):
            current_node = current_node.next
            
            stack.append(current_node)
        
        length = len(stack)
        tmp_next = None
        for item in stack:
            (tmp_next, item.next) = (item, tmp_next) 
        
        return stack[len(stack) - 1]
```

```python
class Solution:
    def tribonacci(self, n: int) -> int:
        trib = []
        trib.append(0)
        trib.append(1)
        trib.append(1)
        if n == 0:
            return 0
        while (len(trib) <= n):
            trib.append(trib[-1] + trib[-2] + trib[-3])
        return trib[-1]
```

```python
class Solution:
    def searchBST(self, root: Optional[TreeNode], val: int) -> Optional[TreeNode]:
        node = root
        if node.val == val:
            return node
        elif node.val > val:
            if root.left is None:
                return None
            return Solution.searchBST(self, root=node.left, val= val)
        else:
            if root.right is None:
                return None
            return Solution.searchBST(self, root=node.right, val= val)
```

```python
class Solution:
    def minCostClimbingStairs(self, cost: List[int]) -> int:
        costs = [0]*len(cost)

        costs[-1] = cost[-1]
        costs[-2] = cost[-2]

        for i in range(len(cost) - 2)[::-1]:
            costs[i] = min(costs[i+1], costs[i+2]) + cost[i]
        
        return min(costs[0], costs[1])
```

```python
class Solution:
    def singleNumber(self, nums: List[int]) -> int:
        result = 0
        for i in nums:
            result = result^i
        return result
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def maxDepth(self, root: Optional[TreeNode]) -> int:
        if root is None:
            return 0
        else:
            return max(Solution.maxDepth(self, root.left), Solution.maxDepth(self, root.right)) + 1
```
